# 2. Data Cleaning

## 2.1 orders.csv

````python
orders_df.info()

# drop the rows with NaN values in 'total_paid':
orders_df = orders_df.dropna(subset=['total_paid'])

# check if rows with NaN values actually got dropped
NaN_remaining = orders_df['total_paid'].isna().sum()

print(f"There are {NaN_remaining} 'NaN' values remaining")
````

````python
# check for duplicates
orders_df.duplicated().sum()
````

````python
# format 'create_date' as datetime
orders_df['created_date'] = pd.to_datetime(orders_df['created_date'])

orders_df.info()
````

````python
# check for range in 'total paid'
orders_df.sort_values(['total_paid'])

# check for range in 'created_date'
orders_df.sort_values(['created_date'])

# check if there are more states than the 5 described in LMS
orders_df['state'].unique()
````

## 2.2 orderlines.csv

````python
# drop the outdated, unused 'product_id'
orderlines_df = orderlines_df.drop(['product_id'], axis=1)
````

````python
# format 'date' as datetime
orderlines_df['date'] = pd.to_datetime(orderlines_df['date'])
````

````python
# replace '.' with '' if it is not a decimal point
orderlines_df['unit_price'] = (orderlines_df['unit_price'].str.replace(r'\.(?=.*\.)', '', regex=True))

# format 'unit_price' as float
orderlines_df['unit_price'] = orderlines_df['unit_price'].astype(float)

# check if there are commas -> NO commas!
# orderlines_df['unit_price'].str.contains(',').sum()
````

Problem with the data: <br>
Orderlines.csv contains<br>
* orders with a 'unit_price' of 0,- € and even a negative value (-119,-, might be due to discounts or typo?)<br>
* and unit_price over 999.99 € are sorted in the 1, 2, ... and so forth, because<br>
