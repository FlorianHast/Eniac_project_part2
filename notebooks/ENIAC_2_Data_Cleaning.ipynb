# 2. Data Cleaning

## 2.1 orders.csv

````python
orders_df.info()

# drop the rows with NaN values in 'total_paid':
orders_df = orders_df.dropna(subset=['total_paid'])

# check if rows with NaN values actually got dropped
NaN_remaining = orders_df['total_paid'].isna().sum()

print(f"There are {NaN_remaining} 'NaN' values remaining")
````

````python
# check for duplicates
orders_df.duplicated().sum()
````

````python
# format 'create_date' as datetime
orders_df['created_date'] = pd.to_datetime(orders_df['created_date'])

orders_df.info()
````

````python
# check for range in 'total paid'
orders_df.sort_values(['total_paid'])

# check for range in 'created_date'
orders_df.sort_values(['created_date'])

# check if there are more states than the 5 described in LMS
orders_df['state'].unique()
````

## 2.2 orderlines.csv

````python
# drop the outdated, unused 'product_id'
orderlines_df = orderlines_df.drop(['product_id'], axis=1)
````

````python
# format 'date' as datetime
orderlines_df['date'] = pd.to_datetime(orderlines_df['date'])
````

````python
# replace '.' with '' if it is not a decimal point
orderlines_df['unit_price'] = (orderlines_df['unit_price'].str.replace(r'\.(?=.*\.)', '', regex=True))

# format 'unit_price' as float
orderlines_df['unit_price'] = orderlines_df['unit_price'].astype(float)

# check if there are commas -> NO commas!
# orderlines_df['unit_price'].str.contains(',').sum()
````

````python
orderlines_df.sort_values(['unit_price'], ascending=False)
````

## 2.3 products.csv

````python
# reset DataFrame
# products.csv
products_df = pd.read_csv('Eniac_project_part2/data/products.csv')
products_df['price'].sort_values().head(5)
````

````python
# replace '.' with '' if it is not a decimal point
products_df['price'] = (products_df['price'].str.replace(r'\.(?=.*\.)', '', regex=True))
products_df['promo_price'] = (products_df['promo_price'].str.replace(r'\.(?=.*\.)', '', regex=True))

# format 'price' as float
products_df['price'] = products_df['price'].astype(float)
products_df['promo_price'] = products_df['promo_price'].astype(float)

# check if there are commas -> NO commas!
# products_df['price'].str.contains(',').sum()

products_df.info()
````

````python
products_df.sort_values(['price'], ascending=False)
````

````python
brands_df
````

# Joined DataFrame

````python
orderlines_df['brand_code'] = orderlines_df['sku'].str[:3]

merged_df = (orderlines_df.merge(orders_df, left_on='id_order', right_on='order_id', how='inner')
              .merge(products_df, on='sku', how='inner')
              .merge(brands_df, left_on = 'brand_code', right_on='short', how='inner')
              )


merged_df = merged_df[['id', 'order_id', 'sku', 'name', 'desc' , 'long', 'product_quantity', 'unit_price', 'price', 'promo_price', 'total_paid', 'state', 'in_stock', 'type', 'date', 'created_date']]


merged_df = merged_df.drop_duplicates()
# merged_df.duplicated().sum()


# merged_df.loc[merged_df['unit_price'].str.startswith('1')].sort_values('unit_price').head(20)

merged_df
````

````python
merged_df['promo_price'] = (merged_df['promo_price'].str.replace(r'\.(?=.*\.)', '', regex=True))
````

````python
merged_df['promo_price'] = merged_df['promo_price'].astype(float)
````

````python
merged_df['factor_promo_unit'] = merged_df['promo_price']/merged_df['unit_price']
````
