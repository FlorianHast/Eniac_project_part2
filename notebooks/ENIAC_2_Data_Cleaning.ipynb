# Joined DataFrame

````python
orderlines_df['brand_code'] = orderlines_df['sku'].str[:3]

merged_df = (orderlines_df.merge(orders_df, left_on='id_order', right_on='order_id', how='inner')
              .merge(products_df, on='sku', how='inner')
              .merge(brands_df, left_on = 'brand_code', right_on='short', how='inner')
              )


merged_df = merged_df[['id', 'order_id', 'sku', 'name', 'desc' , 'long', 'product_quantity', 'unit_price', 'price', 'promo_price', 'total_paid', 'state', 'in_stock', 'type', 'date', 'created_date']]


merged_df = merged_df.drop_duplicates()
# merged_df.duplicated().sum()


# merged_df.loc[merged_df['unit_price'].str.startswith('1')].sort_values('unit_price').head(20)

merged_df
````

````python
# # replace '.' with '' if it is not a decimal point
merged_df['unit_price'] = (merged_df['unit_price'].str.replace(r'\.(?=.*\.)', '', regex=True))

# # format 'unit_price' as float
merged_df['unit_price'] = merged_df['unit_price'].astype(float)
````
